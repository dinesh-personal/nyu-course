```python
import numpy as np
import utils as ut
from sklearn import linear_model as lm
```

```python
X, Y = ut.read_data('data/target_data.csv', x_cols=2)

# It is a good idea to print a few numbers to validate that data has been read correctly 

print(X[0:10])
print(Y[0:10])
```

```python
# It is a good idea to plot the data to see if we can figure out the relationships 
x0_values = [x[0] for x in X]
x1_values = [x[1] for x in X] 
ut.plot_3d(x0_values, x1_values,Y,'D1','D2','Target')
```

```python
#Let us define a generic routine for prediction 
def run_prediction(model, X, Y, label, score_only=False):
    model.fit(X,Y)
    if not score_only:
        print(f'{label} Coefficients are: {model.coef_}')
    print(f'{label} Score is {model.score(X,Y)}')
```

```python
model = lm.LinearRegression()
run_prediction(model, X, Y, 'Least Squares - unscaled')
```

```python
model = lm.Ridge(alpha=0.5)
run_prediction(model, X, Y, 'Ridge')
```

```python
model = lm.Lasso(alpha=0.1)
run_prediction(model, X, Y, 'Lasso')
```
